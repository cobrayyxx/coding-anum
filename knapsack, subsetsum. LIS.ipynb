```python
import numpy as np
def SubsetSum(arr, num):
    #Buat matriks 2 dimensi untuk menyimpan True/False 
    baris = len(arr)+1
    kolom = num+1
    arr2d=numpy.zeros((baris,kolom))
    # Inisiasi kolom pertama adalah 1
    for i in range(baris):
        arr2d[i][0] = 1
    # inisiasi baris pertama adalah 0
    for i in range(1,kolom):
        arr2d[0][i] = 0
    # Lakukan looping untuk pengecekan apakah nilai tersebut nilai yang ada di kolom
    for i in range(1,baris):
        for j in range(1,kolom):
            arr2d[i][j]=arr2d[i-1][j]
            if arr2d[i][j] == False and j>= arr[i-1]:
                arr2d[i][j] = arr2d[i][j] or arr2d[i-1][j-arr[i-1]]
                print("baris ke", i,j)
                print(arr2d)
   # Return True kalau terdapat subset yang berjumlah num dengan cara melihat baris paling akhir dan kolom paling akhir
    return arr2d[baris-1][kolom-1]
# source: https://www.youtube.com/watch?v=34l1kTIQCIA
#source: https://www.youtube.com/watch?v=5td2QH-x5ck
```

# Knapsack


```python
import numpy as np
def knapsack(weight_arr,profit_arr,max_weight):
    # buat array 2D untuk menyimpan hasil penghitungan
    baris = len(weight_arr)+1
    kolom = max_weight+1
    arrV = numpy.zeros((baris,kolom))
    # sisipkan elemen 0  di array weight dan array profit supaya mudah saat penghitungan di dalam array 2d
    weight_arr.insert(0,0) 
    profit_arr.insert(0,0)
    
    #lakukan looping pada array 2D untuk memulai mencari profit tertinggi dengan weight<=max weight
    #looping dilakukan dari 1 supaya baris dan kolom ke-0 cukup diisi dengan 0
    for i in range(1,baris):
        for j in range(1,kolom):
            # jika j-weight_arr[i]<0 maka saat pengambilan elemen arrV akan index out of bound 
            # maka dari itu isi value arrV[i,j] dengan elemen tepat 1 baris diatasnya 
            if j-weight_arr[i]<0:
                arrV[i][j] = arrV[i-1][j]
            else:
                #cari value mana yang mengandung profit paling maks antara elemen tepat satu baris diatasnya
                #dengan elemen satu baris di atasnya namun kolomnya digeser kekiri sebesar weight yang dimiliki 
                #dan dijumlahkan profit yang dimiliki
                arrV[i][j] = max(arrV[i-1][j], arrV[i-1][j-weight_arr[i]]+profit_arr[i])
            print("baris ke", i,j)
            print(arrV)
    # Jika sudah selesai kita lihat baris paling akhir dan kolom paling akhir
    # maka itulah profit maksimal dengan weight<=max_weight       
    return arrV[baris-1][kolom-1]
    
```

```python
knapsack([2,3,4,5],[1,2,5,6],8)
```
